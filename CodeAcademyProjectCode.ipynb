```python
# clean owners dataframe
```

```python
# remove unnecessary columns
drop_columns = ['What is your postal code (zip code)?', 'Submitted At', 'Token']
owners = owners.drop(labels = drop_columns, axis = 1)

# check dataframe
owners.head()
```

```python
# rename columns
column_mapper = {'What is your email?': 'ID',
                'What is your gender?': 'Gender',
                'What is your age?':'Age'}
owners = owners.rename(mapper = column_mapper, axis = 1)

# check dataframe
owners.head()
```

```python
# check data types and # of non-null values in each column
owners.info()
```

```python
# merge dogs and owners dataframe
```

```python
# check how many unique values there are in owners and check data type of ID column
owners['ID'].describe()
```

```python
# convert all owners IDs to string
owners['ID'] = owners.ID.astype('string', errors = 'ignore')

# verify all owners IDs are strings
print(owners['ID'][0:2000])
```

```python
# some owners submitted multiple entries so we need to drop duplicate IDs and assign to variable uniqueowners
uniqueowners = owners['ID'].drop_duplicates()

# assign name to column 
uniqueowners.columns = ['ID']

 # verify all IDs are strings
uniqueowners[:2000]
```

```python
# verify that there are no duplicates
print(uniqueowners.nunique())
print(uniqueowners.count())
```

```python
# merge uniqueowners dataframe with owners using ID column
# using a left merge with uniqueowners on left because we want to maintain all rows in uniqueowners
cleaned_owners = pd.merge(left = uniqueowners, right = owners,
                          left_on = 'ID',
                          right_on = 'ID',
                          how = 'left')
cleaned_owners['ID'] = cleaned_owners['ID'].drop_duplicates()

# verify there are still no duplicate IDs
cleaned_owners['ID'].describe()
```

```python
# merge cleaned_owners dataframe with dogs
# using a left merge with dogs on left because we want to maintain all rows in dogs even if there are duplicate dog names
dogsandowners = pd.merge(left = dogs, right = cleaned_owners,
                         left_on = 'ID',
                         right_on = 'ID',
                         how = 'left')    
dogsandowners.head()
```

```python
# check how many dog names should be in merged dataframe
dogs['Dog name'].describe()
```
